```python
import numpy as np
from PIL import Image
import os
import math
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
image_dir = '/home/chrisjan/project/training/koi_train/images'
image_paths = [image_dir + '/' + filename for filename in os.listdir(image_dir)]
IMAGE_SIZE = (24,16)
```

```python
images = [Image.open(i) for i in image_paths]
for i in range(8, len(images[:33]),8):
    imgs_comb1 = np.hstack((np.asarray(j) for j in images[i-8:i-4]))
    imgs_comb2 = np.hstack((np.asarray(j) for j in images[i-4:i]))
    imgs_comb = np.vstack((imgs_comb1, imgs_comb2))
    imgs_comb = Image.fromarray(imgs_comb)
    width, height = imgs_comb.size
    imgs_comb = imgs_comb.resize([math.floor(width * 0.4), math.floor(height * 0.4)])
    plt.figure(figsize=IMAGE_SIZE)
    plt.imshow(imgs_comb)
```
